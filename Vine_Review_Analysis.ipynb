```python
# import libraries
import pandas as pd
```

```python
# Read vine table
df = pd.read_csv('vine_table.csv')
```

```python
# display dataframe
df
```

```python
# retrieve rows where total votes is equal or greater than 20
total_vote_count = df[df['total_votes'] >= 20]
total_vote_count
```

```python
# Retrieve rows where total vote count where helpful votes divided by total votes is greater than 50%
percentage_of_votes = total_vote_count[(total_vote_count['helpful_votes'] / total_vote_count['total_votes']) >= 0.5]
```

```python
# Retrieve all the rows where a review was written as part of the Vine program (paid),
paid_vine_program = percentage_of_votes[percentage_of_votes['vine'] == 'Y']
paid_vine_program
```

```python
# Retrieve all the rows where a review was not written as part of the Vine program (not paid),
unpaid_vine_program = percentage_of_votes[percentage_of_votes['vine'] == 'N']
```

```python
# List of total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid)
total_reviews = paid_vine_program['total_votes'].sum()
number_five_star = (paid_vine_program['star_rating'] == 5).sum()
five_star_percentage = round((number_five_star / (paid_vine_program['star_rating']).count())* 100)
```

```python
total_unpaid_reviews = unpaid_vine_program['total_votes'].sum()
number_unpaid_five_star = (unpaid_vine_program['star_rating'] == 5).sum()
unpaid_five_star_percentage = round((number_unpaid_five_star / (unpaid_vine_program['star_rating']).count())* 100)
```

```python
new_data = {'Vine':[total_reviews,number_five_star,five_star_percentage],
            'Non-Vine': [total_unpaid_reviews,number_unpaid_five_star,unpaid_five_star_percentage]    
}

Final_analysis =pd.DataFrame.from_dict(new_data,orient ='index', columns=['Total number of reviews',
                                        'Number of 5-star reviews',
                                        'Percentage of 5-star reviews']
)
```
